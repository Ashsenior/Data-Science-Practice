```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# recomended method 

x = [1,2,4,6]
y = [7,4,8,0]

# fig is figure and ax stands for axes
fig, ax = plt.subplots(figsize=(5,5))

# adding labels and title 
ax.set(title='New data',xlabel='x axis',ylabel='y axis')

# plotting the data
ax.plot(x,y);
```

```python
# scatter plot and bar plot
fig , ax = plt.subplots()
k = np.linspace(0,10,70)
ax.set(title='Parbola',xlabel='x',ylabel='x squared')
ax.plot(k,np.exp(k));
```

```python
fig , ax = plt.subplots()
ax.set(xlabel='x',ylabel='sin(x)',title='Sine graph')
ax.scatter(k,np.sin(k));
```

```python
nut_butter_sales = {'almond':10,'peanut':8,'cashew':14}
fig , ax = plt.subplots()
ax.set(title='Nut butter sales',xlabel='Butter type',ylabel='Price')
ax.bar(nut_butter_sales.keys(),nut_butter_sales.values());
```

```python
# multiple subplots in single image
fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
# plotting each axes
ax0.set(title='Line plot')
ax0.plot(k,k**2);

ax1.set(title='Scatter plot')
ax1.scatter(k,np.sin(k));

ax2.set(title='Bar plot')
ax2.bar(nut_butter_sales.keys(),nut_butter_sales.values());

ax3.set(title='Histogram')
ax3.hist(np.random.randn(1000));
```

```python
import pandas as pd
pd.set_option('max_rows',6)
data = pd.read_csv("new_data.csv")
data.head()
```

```python
# first method
fig, ax = plt.subplots()
cut_data = data.loc[0:200]
cut_data['Short_form'] = cut_data['Country'].str[0:2]
ax.set(title='Fortune 2000')
ax.plot(cut_data.Rank,cut_data.Sales);
```

```python
cut_data
```

```python
# second method
country_wise = data.groupby(by='Country').count()
country_wise.sort_values(by='Name',ascending=False,inplace=True)
top5 = country_wise.head(10)
```

```python
top5.plot(kind='pie',y='Name');
```

```python
cut_data.loc[0:100].plot(figsize=(10,12),kind='hist',subplots=True,bins=50);
```

```python
fig , ax = plt.subplots(figsize=(10,10))
scatter = ax.scatter(x=cut_data.Short_form,y=cut_data.Sales,c=cut_data.Rank);
ax.set(title='Country wise ranking of companies',xlabel='Conutry',ylabel='Sales ')
ax.legend(*scatter.legend_elements(),title='Rank');
ax.axhline(cut_data.Sales.mean(),linestyle='--');
```

```python
cut_data.head()
```

```python
fig , (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter = ax0.scatter(x=cut_data.Short_form,y=cut_data.Sales,c=cut_data.Rank,cmap='plasma');
ax0.set(title='Country wise sales of companies',xlabel='Country',ylabel='Sales');
ax0.legend(*scatter.legend_elements(),title='Rank');
ax0.axhline(cut_data.Sales.mean(),linestyle='--');

scatter2 = ax1.scatter(x=cut_data.Short_form,y=cut_data.Market_value,c=cut_data.Rank,cmap='inferno');
ax1.set(title='Country wise Market value of companies',xlabel='Country',ylabel='Market Value');
ax1.legend(*scatter2.legend_elements(),title='Rank');
ax1.axhline(cut_data.Market_value.mean(),linestyle='--');

fig.suptitle("Fortune 2000 companies analysis",fontsize=15);
```
