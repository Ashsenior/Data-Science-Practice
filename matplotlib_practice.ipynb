```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# recomended method 

x = [1,2,4,6]
y = [7,4,8,0]

# fig is figure and ax stands for axes
fig, ax = plt.subplots(figsize=(5,5))

# adding labels and title 
ax.set(title='New data',xlabel='x axis',ylabel='y axis')

# plotting the data
ax.plot(x,y);
```

```python
# scatter plot and bar plot
fig , ax = plt.subplots()
k = np.linspace(0,10,70)
ax.set(title='Parbola',xlabel='x',ylabel='x squared')
ax.plot(k,np.exp(k));
```

```python
fig , ax = plt.subplots()
ax.set(xlabel='x',ylabel='sin(x)',title='Sine graph')
ax.scatter(k,np.sin(k));
```

```python
nut_butter_sales = {'almond':10,'peanut':8,'cashew':14}
fig , ax = plt.subplots()
ax.set(title='Nut butter sales',xlabel='Butter type',ylabel='Price')
ax.bar(nut_butter_sales.keys(),nut_butter_sales.values());
```

```python
# multiple subplots in single image
fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
# plotting each axes
ax0.set(title='Line plot')
ax0.plot(k,k**2);

ax1.set(title='Scatter plot')
ax1.scatter(k,np.sin(k));

ax2.set(title='Bar plot')
ax2.bar(nut_butter_sales.keys(),nut_butter_sales.values());

ax3.set(title='Histogram')
ax3.hist(np.random.randn(1000));
```

```python
import pandas as pd
pd.set_option('max_rows',6)
data = pd.read_csv("new_data.csv")
data.head()
```

```python
# first method
fig, ax = plt.subplots()
cut_data = data.loc[0:200]
ax.set(title='Fortune 2000')
ax.plot(cut_data.Rank,cut_data.Sales);
```

```python
# second method
country_wise = data.groupby(by='Country').count()
country_wise.sort_values(by='Name',ascending=False,inplace=True)
top5 = country_wise.head(10)
```

```python
top5.plot(kind='pie');
```
